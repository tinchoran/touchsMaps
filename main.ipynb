```python
# %%
import pandas as pd
from statsbombpy import sb
import matplotlib.pyplot as plt
from mplsoccer import VerticalPitch, Pitch

partido = sb.events(match_id=3869685)

messi = partido[partido["player"] == "Lionel Andrs Messi Cuccittini"]
name = set()
for index, row in messi.iterrows():
    jugador = row["player"]
    name.add((jugador))

name_list = list(name)

messi[["x","y"]] = messi["location"].apply(pd.Series)

goals = messi[messi["shot_outcome"] == "Goal"]

messi[["x", "y", "location"]]

fig, ax = plt.subplots(figsize=(16,9))
fig.patch.set_facecolor('none')
pitch = Pitch(
    pitch_color='grass', 
    line_color='white', 
    pitch_type="statsbomb",
    stripe=True
)
# Ajustar lmites y mrgenes para eliminar el relleno blanco


pitch.scatter(messi["x"], messi["y"], ax = ax, c='#f7f7f7', edgecolor="#6b0002")
# Dibujar los goles con un color diferente
pitch.scatter(goals["x"], goals["y"], ax=ax, c='blue', edgecolor="white", label="Goles")
ax.legend(facecolor='#22312b', edgecolor='None', fontsize=15, loc='upper left', fancybox=False)
pitch.draw(ax = ax)

```
